```python
import numpy
import pandas as pd
```

```python
data = pd.read_csv("C:/Users/Shivangi/Downloads/IndianCitiesLatLong.csv")
```

```python
data.head()
```

```python
c = input("enter city name")
```

```python
ind = data[data['City']==c].index.values

for i in range(len(data)):
    closest = []
    dist = (data["Latitude"]-data["Latitude"][ind[0]])**2 - (data["Longitude"]-data["Longitude"][ind[0]])**2
    closest= dist.tolist()
print(closest)
```

```python
new_list = []
for i in closest:
    if i<0:
        new_list.append(i * -1)
    else:
        new_list.append(i)
        
print(new_list)
```

```python
def merge(list1, list2): 
      
    merged_list = [(list1[i], list2[i]) for i in range(0, len(list1))] 
    return merged_list 
city_dist =merge(data["City"], new_list) 
print(city_dist)
```

```python
sorted_list= sorted(city_dist,key=lambda x: x[1])
print(sorted_list)
```
